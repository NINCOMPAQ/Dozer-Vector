```python
import pandas as pd
import ast
import numpy as np
import torch

df = pd.read_csv("CHIBike.geo")
```

```python
df['centers'] = df['coordinates'].copy()
def getCenter(strCoords):
    parsed_coords = np.array(ast.literal_eval(strCoords)[0])
    return parsed_coords.mean(axis=0)
df['centers'] = df['centers'].apply(func = getCenter).copy()
```

```python
#vec_field = [[0]*270]*270
vec_field = [[0]*270 for _ in range(270)]
for index, row in df.iterrows():
    for index1, row1 in df.iterrows():
        delta_x = row1['centers'][0] - row['centers'][0]
        delta_y = row1['centers'][1] - row['centers'][1]
        vec_field[index][index1]= np.array([delta_x,delta_y])
        
vec_field
```
